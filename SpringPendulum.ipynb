```python
import numpy as np
import matplotlib.pyplot as plt

m = 1
k = 50
g = 10
l = 1
pi = np.pi

endt = 4
```

```python
radi = [l, 0, 0]
thet = [pi/3, 0, 0]

dt = 0.00001

rd = [radi[0]]
th = [thet[0]]

rdv = [radi[1]]
thv = [thet[1]]

time = [0]

while(thet[0] > 0):
#while(time[-1] < endt):
    radi[2] = (m*radi[0]*(thet[1]**2) - k*(radi[0]-l) + m*g*np.cos(thet[0])) / m
    thet[2] = (-m*g*radi[0]*np.sin(thet[0]) - 2*m*radi[0]*radi[1]*thet[1]) / (m*(radi[0]**2))
    
    
    radi[0] = radi[0] + radi[1]*dt + 0.5*radi[2]*(dt**2)
    thet[0] = thet[0] + thet[1]*dt + 0.5*thet[2]*(dt**2)
    
    radi[1] = radi[1] + radi[2]*dt
    thet[1] = thet[1] + thet[2]*dt
    
    rd.append(radi[0])
    th.append(thet[0])
    rdv.append(radi[1])
    thv.append(thet[1])
    
    time.append(time[-1]+dt)
```

```python
    plt.figure()
    plt.plot(time, rd, label="m1")
    plt.title('radius')
    plt.xlabel('Time [s]')
    plt.ylabel('radial distance')
    plt.legend()
```

```python
    plt.figure()
    plt.plot(time, th, label="m1")
    plt.title('angular displacement')
    plt.xlabel('Time [s]')
    plt.ylabel('radial distance')
    plt.legend()
```

```python
x = []
y = []

v = []

H = []

for i in range(len(rd)):
    x.append(rd[i]*np.sin(th[i]))
    y.append(-rd[i]*np.cos(th[i]))
    
    v.append(np.sqrt(rdv[i]**2 + (rd[i]*thv[i])**2))
```

```python
plt.figure()
plt.plot(x, y, label="m1")
plt.title('trajectory')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```

```python
plt.figure()
plt.plot(time, v, label="m1")
plt.title('velocity')
plt.xlabel('time [s]')
plt.ylabel('velocity')
plt.legend()
```

```python
H = []
for i in range(len(rd)):
    H.append(0.5*m*v[i]**2 + 0.5*k*(rd[i]-l)**2 - m*g*rd[i]*np.cos(th[i]))

plt.figure()
plt.plot(time, H, label="m1")
plt.title('hamiltonian')
plt.xlabel('time [s]')
plt.ylabel('hamiltonian')
plt.legend()
```
